```python
user_genre = pd.read_csv ('user_genre.csv.gz', compression='gzip')
```

```python
user_genre
```

```python
# Convertir la columna 'genres' a una lista si aún no está en ese formato
#user_genre['genres'] = user_genre['genres'].apply(eval)

# Usar explode para convertir listas en filas individuales
exploded_df = user_genre.explode('genres')

# Encontrar el usuario que más jugó y el año que más se jugó para cada género
agg_df = exploded_df.groupby('genres').agg({
    'user_id': pd.Series.idxmax,
    'year': 'idxmax'
})

# Filtrar el DataFrame original utilizando los índices encontrados
result_df = exploded_df.loc[agg_df['user_id'], ['genres', 'user_id', 'year']]

# Renombrar las columnas según tus especificaciones
result_df.columns = ['genres', 'nombre', 'max_year']

# Usar explode nuevamente en result_df para mantener el formato deseado en 'genres'
result_df['genres'] = result_df['genres'].apply(lambda x: [x])

# Imprimir el resultado
result_df

```
